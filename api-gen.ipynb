```python
# 1. Clone the Web UI repository
!git clone -q https://github.com/AUTOMATIC1111/stable-diffusion-webui.git

# 2. Create directory for models
!mkdir -p stable-diffusion-webui/models/Stable-diffusion

# 3. Install dependencies
!pip install -q huggingface_hub

# 4. Download a high-quality Stable Diffusion model and its VAE
from huggingface_hub import hf_hub_download
import os

print("Downloading Stable Diffusion models (this may take a few minutes)...")

# Download the main model (Realistic Vision v4.0 with inpainting)
main_model_path = hf_hub_download(
    repo_id="SG161222/Realistic_Vision_V4.0",
    filename="Realistic_Vision_V4.0-inpainting.safetensors",
    local_dir="stable-diffusion-webui/models/Stable-diffusion/"
)

# Download the corresponding VAE for better quality
vae_model_path = hf_hub_download(
    repo_id="SG161222/Realistic_Vision_V4.0",
    filename="realvis4-vae-ft-mse-840000-ema-pruned.safetensors",
    local_dir="stable-diffusion-webui/models/Stable-diffusion/"
)

# Rename the VAE to match the model filename pattern so it loads automatically
new_vae_path = main_model_path.replace(".safetensors", ".vae.safetensors")
os.rename(vae_model_path, new_vae_path)

print("✅ Model and VAE download complete.")
```
